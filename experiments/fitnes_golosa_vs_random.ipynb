```python
import numpy as np
from scipy import stats
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
from numpy import genfromtxt
```

```python
data = open("fitness_golosa_random", "r")
vec = []
for i in iter(data.readline, ''):
    vec.append(float(i[:-1]))
```

```python
vec_np = np.array(vec)
print np.percentile(vec_np, 25)
print(stats.describe(vec_np))
```

```python
sns.kdeplot(vec_np)
plt.xlim(.7,1)
plt.axvline(vec_np.mean(), linestyle='--', c='g')
plt.axvline(np.percentile(vec_np,25), linestyle='--', c='r')
plt.axvline(np.percentile(vec_np,75), linestyle='--', c='r')
plt.show()

plt.hist(vec_np, 100, normed=1, histtype='step', cumulative=True)
plt.axhline(1, linestyle='--', c='g')
plt.xlim(.7,1)
plt.show()
```

```python
def parse_gen_data(route):
    data_random_fitness = genfromtxt(route,delimiter=',')
    data_random_fitness_generations = np.array(data_random_fitness[:,0],dtype=np.int)
    data_random_fitness_fitness = np.array(data_random_fitness[:,1],dtype=np.float)
    return data_random_fitness_generations, data_random_fitness_fitness
```

```python
data_random_fitness_generations, data_random_fitness_fitness = \
    parse_gen_data("genetic/random_fitness_100_50_0.005")
data_random_fitness_generations_2, data_random_fitness_fitness_2 = \
    parse_gen_data("genetic/random_fitness_100_25_0.005")
data_random_fitness_generations_3, data_random_fitness_fitness_3 = \
    parse_gen_data("genetic/random_fitness_100_100_0.005")
```

```python
plt.plot(data_random_fitness_generations_3,data_random_fitness_fitness_3, "g-", label="100")
plt.plot(data_random_fitness_generations,data_random_fitness_fitness, "r-", label="50")
plt.plot(data_random_fitness_generations_2,data_random_fitness_fitness_2, "b-", label="25")
plt.legend(loc='best')
plt.ylim(0,1)
plt.show()
```
