```python
from __future__ import unicode_literals
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
```

```python
data = pd.read_csv('save/grid/grasp_P20.csv')
data['cum_champs'] = data[" championships"].cumsum()
```

```python
# plt.plot(data.index, data['fitness'])
# plt.show()
# plt.plot(data['cum_champs'], data['fitness'])
# plt.xlabel("Cantidad de partidos jugados")
# plt.ylabel("Fitness")
# plt.show()

fits = []
for idx in data.index:
    for i in range(data.loc[idx," championships"]):
        fits.append(data.loc[idx,"fitness"])
print len(fits)
plt.plot(fits)
plt.xlabel("Cantidad de partidos jugados")
plt.ylabel("Fitness")
plt.savefig('plots/grid_grasp_P20.pdf')
plt.show()
```

```python
serie_fit = pd.Series(fits)
print serie_fit.shape
print serie_fit[serie_fit>0.97].shape
```

```python
fitness20 = [0] * 21
currentP = 0 
for idx in data.index:
    if data.loc[idx," championships"] >= currentP:
        for i in range (currentP, data.loc[idx," championships"]+1):
            fitness20[i] = data.loc[idx, "fitness"]
        currentP = data.loc[idx," championships"]+1;

f, ax = plt.subplots(1)
ax.set_ylim(ymin=0, ymax=1)
plt.plot(fitness20)
plt.show(f)
```
